```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.options.display.max_rows = None
pd.options.display.max_columns = None

read_file = pd.read_excel('C:/Users/HP/Desktop/Churn Prediction/customer churn data_usecase2_Hackathon.xlsx')
read_file.to_csv("BankData.csv",index = None)
df = pd.read_csv("BankData.csv")
df.head(1)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')  # This downloads the Open Multilingual Wordnet
nltk.download('stopwords')  # This is for stopwords
nltk.download('punkt')  # This is for tokenization
nltk.download('vader_lexicon')
# Preprocess the feedback text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove special characters and digits
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Tokenization
    words = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    return " ".join(words)

# Apply the preprocessing function to the 'Customer Feedback' column
df['Processed_Feedback'] = df['Customer Feedback'].apply(preprocess_text)

# View the processed feedback
df[['Customer Feedback', 'Processed_Feedback']].head(5)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define a function to analyze the sentiment
def analyze_sentiment(text):
    sentiment_score = sid.polarity_scores(text)
    if sentiment_score['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_score['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to the processed feedback
df['Sentiment'] = df['Processed_Feedback'].apply(analyze_sentiment)

# View the results
df[['Processed_Feedback', 'Sentiment']].head(5)
```

```python
import spacy

# Load the spaCy model for NER (Named Entity Recognition)
nlp = spacy.load("en_core_web_sm")

# Define a function to extract entities from the feedback
def extract_entities(text):
    doc = nlp(text)
    entities = [(entity.text, entity.label_) for entity in doc.ents]
    return entities

# Apply the entity extraction function to the processed feedback
df['Entities'] = df['Processed_Feedback'].apply(extract_entities)

# View the results
df[['Processed_Feedback', 'Entities']].head(5)
```
